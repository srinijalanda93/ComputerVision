```python
# Demonstrate frame extraction from the video and display the color components of the
# images.
# Use OpenCV and Matplotlib in python and perform the following
# 1. Load the video using OpenCV (cv2.VideoCapture).
# 2. Extract frames at specific intervals.
# 3. Split each frame into its red, green, and blue components.
# 4. Display the original frame alongside its individual color channels.
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
video=cv2.VideoCapture('video.mkv')
print("the type of the variable it is ",type(video))
```

```python
# 2.extracting frames at specific intervals
frame_interval=int(input("Enter your frame Interval:"))
# frame_interval = 30  # extracting the 30th frame or any nth frame
frame_count = 0 #set the frame_count inital zero
```

```python
# 3.reading frames from the video
ret, frame = video.read()
print(ret,frame) # boolean value,3d array
print(frame.shape)  #720 is the height, 1280 is the width, and 3 represents the color channels, such as RGB. dimensionality of image
```

```python
while video.isOpened():
    ret, frame = video.read()  # Read a frame
    if not ret:
        print("OOPS! SOMETHING WENT WRONG WITH VIDEO")
        break  # Break if the video ends
    
    if frame_count % frame_interval == 0:
        # convert frame from BGR to RGB (Matplotlib expects RGB format)
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Split into Red, Green, and Blue channels
        r_channel = frame_rgb.copy()
        r_channel[:, :,1:3:1] = 0  # Keep only the Red channel
        
        g_channel = frame_rgb.copy()
        g_channel[:, :, [0, 2]] = 0  # Keep only the Green channel
        
        b_channel = frame_rgb.copy()
        b_channel[:, :, :2] = 0  # Keep only the Blue channel

        # Display original frame and its color components
        fig, axes = plt.subplots(1, 4, figsize=(12, 4))
        axes[0].imshow(frame_rgb)
        axes[0].set_title("Original")
        axes[1].imshow(r_channel)
        axes[1].set_title("Red Channel")
        axes[2].imshow(g_channel)
        axes[2].set_title("Green Channel")
        axes[3].imshow(b_channel)
        axes[3].set_title("Blue Channel")

        for ax in axes:
            ax.axis("off")

        plt.show()

    frame_count += 1

# Release the video capture object
video.release()
```
