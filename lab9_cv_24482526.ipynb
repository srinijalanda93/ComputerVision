```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def apply_sobel(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
    sobel_combined = cv2.magnitude(sobelx, sobely)
    return sobelx, sobely, sobel_combined

def apply_prewitt(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    kernelx = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
    kernely = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
    prewittx = cv2.filter2D(gray, -1, kernelx)
    prewitty = cv2.filter2D(gray, -1, kernely)
    prewitt_combined = cv2.magnitude(prewittx.astype(float), prewitty.astype(float))
    return prewittx, prewitty, prewitt_combined

def apply_roberts(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    kernelx = np.array([[1, 0], [0, -1]])
    kernely = np.array([[0, 1], [-1, 0]])
    robertsx = cv2.filter2D(gray, -1, kernelx)
    robertsy = cv2.filter2D(gray, -1, kernely)
    roberts_combined = cv2.magnitude(robertsx.astype(float), robertsy.astype(float))
    return robertsx, robertsy, roberts_combined

def apply_laplacian(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    laplacian = cv2.Laplacian(gray, cv2.CV_64F)
    return laplacian

def apply_canny(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    canny_edges = cv2.Canny(gray, 50, 150)
    return canny_edges


img = cv2.imread("/content/set-2-program-2.jpg")
sobel_x, sobel_y, sobel_combined = apply_sobel(img)
plt.imshow(sobel_combined, cmap='gray')
plt.show()
```
