```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
#load the images using cv2
img=cv2.imread('Program-1.jpg',cv2.IMREAD_COLOR)
#READ THE IMAGE
plt.imshow(img)
img.shape

height,width=img.shape[:2]
print(height,width)
```

```python
#define the interpolation along with the resize 

def Interpolation(image_path, scale=2):
   
    img = cv2.imread(image_path, cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  
    
    heigth, width = img.shape[:2]
    new_size = (width * scale, height * scale)
    
    #define interpolation
    nearest = cv2.resize(img, new_size, interpolation=cv2.INTER_NEAREST)
    bilinear = cv2.resize(img, new_size, interpolation=cv2.INTER_LINEAR) #4nnp+smoothin+bluring(intor)
    bicubic = cv2.resize(img, new_size, interpolation=cv2.INTER_CUBIC) #16nnp+smoothing+artifact
    lanczos = cv2.resize(img, new_size, interpolation=cv2.INTER_LANCZOS4) #sinc+resampling+Sharpening
    
    #display using the list  title
    titles = ['Original', 'Nearest Neighbor', 'Bilinear', 'Bicubic', 'Lanczos']
    images = [img, nearest, bilinear, bicubic, lanczos]
    
    plt.figure(figsize=(15, 8))
    for x in range(5):
        plt.subplot(2, 3, x + 1)
        plt.imshow(images[x])
        plt.title(titles[x])
    plt.show()
```

```python
#call the function
#load the images
img=cv2.imread('Program-1.jpg',cv2.IMREAD_COLOR)
#READ THE IMAGE
plt.imshow(img)
Interpolation('Program-1.jpg', scale=2)
```
